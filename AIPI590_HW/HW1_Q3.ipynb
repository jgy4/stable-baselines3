```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from stable_baselines3 import A2C
from stable_baselines3.common.evaluation import evaluate_policy
```

## N-Step Advantage Model

```python
env = gym.make('CartPole-v1')
eval_env = gym.make('CartPole-v1')
model = A2C('MlpPolicy', env, verbose=1, n_step=True)
model.learn(total_timesteps=int(1e5))
```

```python
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

## Vanilla Model

```python
env = gym.make('CartPole-v1')
eval_env = gym.make('CartPole-v1')
model = A2C('MlpPolicy', env, verbose=1, n_step=False, gae_lambda=0)
model.learn(total_timesteps=int(1e5))
```

```python
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

## Monte Carlo

```python
env = gym.make('CartPole-v1')
eval_env = gym.make('CartPole-v1')
model = A2C('MlpPolicy', env, verbose=1, n_step=False, gae_lambda=1)
model.learn(total_timesteps=int(1e5))
```

```python
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

### Default GAE Without N-Step

```python
env = gym.make('CartPole-v1')
eval_env = gym.make('CartPole-v1')
model = A2C('MlpPolicy', env, verbose=1, n_step=False)
model.learn(total_timesteps=int(1e5))
```

```python
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)

print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

In comparing the models we found the following results:

*   N-Step Mean Reward: 500
*   Vanilla Mean Reward: 70
*   Monte Carlo Mean Reward: 263.5
*   GAE Mean Reward: 500

The N-Step and GAE models performed the best with the highest mean rewards. The Monte Carlo model performed slightly worse with a mean reward of 263.5. The Vanilla model performed the worst with a mean reward of only 70.
